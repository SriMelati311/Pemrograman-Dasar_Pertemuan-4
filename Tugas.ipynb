```python
import datetime

print("="*40)
print("SELAMAT DATANG DI KAFE BORLAB")
print("="*40)
print("Menu Makanan & Minuman")
print("1. Kopi Hitam    - Rp 12.000")
print("2. Teh Manis     - Rp 10.000")
print("3. Roti Bakar    - Rp 18.000")
print("4. Mie Instan    - Rp 10.000")
print("-"*40)

# Input jumlah tiap menu
print("Pesanan Anda:")
kopi = int(input("Kopi Hitam    : "))
teh = int(input("Teh Manis     : "))
roti = int(input("Roti Bakar    : "))
mie  = int(input("Mie Instan    : "))

# Hitung total harga
harga_kopi = kopi * 12000
harga_teh  = teh * 10000
harga_roti = roti * 18000
harga_mie  = mie * 10000

total_harga = harga_kopi + harga_teh + harga_roti + harga_mie

# Hitung diskon
diskon = 0
if total_harga > 50000:
    diskon = 0.70
elif total_harga > 30000:
    diskon = 0.50
elif total_harga > 20000:
    diskon = 0.30

potongan = int(total_harga * diskon)
total_bayar = total_harga - potongan

# Cetak struk
print("\n" + "="*40)
print("         STRUK PEMBAYARAN")
print("Tanggal :", datetime.datetime.now().strftime("%d-%m-%Y %H:%M:%S"))
print("="*40)
print(f"{'Nama Item':<15}{'Qty':<5}{'Harga':<10}{'Subtotal'}")
print("-"*40)
if kopi > 0:
  print(f"{'Kopi Hitam':<15}{kopi:<5}Rp 12000   Rp {harga_kopi}")
if teh  > 0:
  print(f"{'Teh Manis':<15}{teh:<5}Rp  10000   Rp {harga_teh}")
if roti > 0:
  print(f"{'Roti Bakar':<15}{roti:<5}Rp 18000   Rp {harga_roti}")
if mie  > 0:
  print(f"{'Mie Instan':<15}{mie:<5}Rp 10000   Rp {harga_mie}")
print("-"*40)
print(f"{'Total Harga':<16}: Rp {total_harga}")
print(f"{'Diskon':<16}: {int(diskon*100)}% (Rp {potongan})")
print(f"{'Total Bayar':<16}: Rp {int(total_bayar)}")
print("="*40)

# Pembayaran
uang_dibayar = int(input("Uang Pembayaran : Rp "))
if uang_dibayar >= total_bayar:
    kembalian = uang_dibayar - total_bayar
    print(f"{'Kembalian':<16}: Rp {kembalian}")
    print("="*40)
    print("Terima kasih telah berbelanja di KAFE BORLAB!")
    print("   Semoga hari Anda menyenangkan ")
else:
    kekurangan = total_bayar - uang_dibayar
    print(f"\nMaaf, uang Anda kurang Rp {kekurangan}.")
```
